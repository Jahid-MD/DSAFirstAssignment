```python
%%javascript
//Q2. Write a program to reverse an array in place? In place means you cannot create a new array. 
// You have to update the original array.

class Stack{
    constructor(){
        this.items=[];
        this.length=0;
    }
    // push function
    push(element)
    {
        // Inserting element into the items
        this.items.push(element);
        this.length=this.length+1;
    }
    // pop function
    pop()
    {
        // return top most element in the stack
        // and removes it from the stack
        // -1 if stack is empty
        if (this.items.length == 0)
            return -1;
        this.length=this.length-1;
        return this.items.pop();
    }
    // peek function
    peek()
    {
        // return the top most element from the stack
        // but does'nt delete it.
        return this.items[this.items.length - 1];
    }
    printStack()
    {
        var str = "";
        for (let i = 0; i < this.items.length; i++)
            str += this.items[i] + " ";
        return str;
    }
   
}
var arr=[0,1,2,3,4]
var stack1= new Stack();

//inserting arr elemnts in the stack.items
for(let i=0;i<arr.length;i++){
    stack1.push(arr[i]) 
}

arr=[]//empting array

const stackLength=stack1.length
//inserting elements back into arr in reverse order
for(let i=0;i<stackLength;i++){
    arr.push(stack1.pop())
}
console.log(arr)
```
